### 4. A block

Print out a grid of 0s and 1s, with width and height equal to n.

n = 3

When n=3, output:<br>
010<br>
101<br>
010<br>

When n=5, output:<br>
01010<br>
10101<br>
01010<br>
10101<br>
01010<br>

```python
#Try here
```

##### Solution

```python
n = 5
for i in range(n):
    for j in range(n):
        if i % 2 ==0:
          if j % 2 == 0:
            print("0",end="")
          else:
            print("1",end="")
        else:
           if j % 2 == 0:
            print("1",end="")
           else:
            print("0",end="")
    print()
```

### 5. A block

3n+1 Problem
Given the following operation:

Take an input x,

If x is divisible by 2, output x/2 else output 3x+1. Repeat this process until you reach 1.

Your task is to find out how many steps you need until the number reaches 1.

**Example 1:**

Input: 10<br>
Output: 6

Reason:<br>
10 --- 5 --- 16 --- 8 --- 4 --- 2 --- 1

**Example 2**

Input: 7<br>
Output: 16

Reason:<br>
7 --- 22 --- 11 --- 34 --- 17 --- 52 --- 26 --- 13 --- 40 --- 20 --- 10 --- 5 --- 16 --- 8 --- 4 --- 2 --- 1


#####Solution

```python
def Solution(x):

    step = 0
    result = x
    reason = [x]

    while result>1:

     if result % 2 == 0:
         result = result/2
     else:
         result = 3*result+1

     step += 1
     reason.append(result)

    print(f'{step}')

    for i in reason:
        print(i,end=" --- ")

Solution(7)
```
