#### 3e.3 Set Operation

```python
students_like_swimming = {1, 2, 4}
students_like_running = {1, 3, 5}
students_like_dancing = {2, 3, 6}

students_like_swimming_or_dancing = students_like_swimming.union(students_like_dancing)
print(students_like_swimming_or_dancing)
# Expected output: {1,2,3,4,6}

students_like_swimming_and_dancing = students_like_swimming.intersection(students_like_dancing)
print(students_like_swimming_and_dancing)
# Expected output: {2}

students_like_swimming_but_not_dancing = students_like_swimming - students_like_dancing
print(students_like_swimming_but_not_dancing)
# Expected output: {1,4}
```

### 5b. Looping Dictionary

#### 5b.1 Loop a dictionary with the .items(), .keys() and .values()

```python
# Example

biggest_brands = {
    "Apple": 2153,
    "Google": 1170,
    "Microsoft": 1165,
    "Coca-Cola": 230.94,
    "Amazon": 1161
}
```

```python
# Looping the keys
for key in biggest_brands:
    print(key)

for key in biggest_brands.keys():
    print(key)

# List Comprehension
print([k for k in biggest_brands])
print([k for k in biggest_brands.keys()])
```

```python
# Looping the values

for value in biggest_brands.values():
    print(value)

# List comprehension version
[v for v in biggest_brands.values()]
```

```python
# Looping through both keys and values

for company, value in biggest_brands.items():
    print(f"{company} has a value of {value} billion dollars.")
```
