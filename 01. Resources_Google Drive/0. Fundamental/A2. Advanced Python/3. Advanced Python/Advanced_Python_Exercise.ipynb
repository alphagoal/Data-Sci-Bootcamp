#### 1a.2 Datetime arithmetic

````python
dob_albert_einstein = datetime(1874, 3, 14)
current_dt = datetime.now()
ten_days_duration = timedelta(days=10)
````

````python
# Difference between two datetime can be represented by timedelta
time_lapsed_since_albert_born = current_dt - dob_albert_einstein
print(time_lapsed_since_albert_born)
````

````python
# Adding two datetime does not make any sense
dob_albert_einstein + current_dt
````

````python
# Adding timedelta to datetime
ten_days_later  = current_dt + ten_days_duration
ten_days_ago  = current_dt - ten_days_duration
print(ten_days_later)
print(ten_days_ago)
````

````python
# Add/subtract two timedelta
twenty_days = ten_days_duration + ten_days_duration
zero_days = ten_days_duration - ten_days_duration
print(twenty_days)
print(zero_days)
````

#### 5a.2 Inplace vs non-inplace operations

````python
clinic_df = pd.DataFrame([
        ['January', 100, 100, 23, 100],
        ['February', 51, 45, 145, 45],
        ['March', 81, 96, 65, 96],
        ['April', 80, 80, 54, 180],
        ['May', 51, 54, 54, 154],
        ['June', 112, 109, 79, 129]
    ],
    columns=['month', 'clinic_east','clinic_north', 'clinic_south','clinic_west'])


# Since the operation mutate the dataframe, let's create new copy
df_copy = clinic_df.copy()
df_copy
````

#### 6b.4 Time series Group By

````python
temperature_record = pd.DataFrame([
    ["1988-04-07 00:00", 30],
    ["1988-04-07 12:00", 29],
    ["1988-04-08 00:00", 30],
    ["1988-04-09 00:00", 29],
    ["1988-04-09 12:00", 28],
    ["1988-04-10 03:00", 31],
], columns=['time', 'reading'])

# to_datetime() is a short cut of .apply() + .strptime()
temperature_record.index = pd.to_datetime(temperature_record['time'], format="%Y-%m-%d %H:%M")

# Now the index is a DateTimeIndex - A special one for datetime

# Then use .sample() as a group-by
resampled = temperature_record.resample("1d")['reading'].mean()

resampled
````

##### **Solution**

````python
class Circle:

  pi = 3.1415926536

  def __init__(self,diameter):
    self.diameter = diameter
    self.circumference = self.pi*diameter

  def num_of_rolls(self,distance):
    roll = distance / self.circumference
    print(f"The circle rolled {roll} rotations")

  def __repr__(self):
    return(f"A circle with diameter: {self.diameter}, circumference:{self.circumference}.")

  def __gt__(self,other):
    return self.diameter > other.diameter

circle_a = Circle(10)
circle_b = Circle(11)

if circle_a > circle_b:
    print("Circle a is larger")
else:
    print("Circle b is larger.")
````
