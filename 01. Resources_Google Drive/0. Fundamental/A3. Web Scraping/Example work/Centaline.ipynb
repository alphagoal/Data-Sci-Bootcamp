```python
import requests
```

```python
url = "https://hk.centanet.com/findproperty/api/Post/Search"

body = {
    "postType":"Rent",
    "sort":"Ranking",
    "order":"Ascending",
    "size": 24,
    "displayTextStyle": "WebResultList",
    "offset": 0,
    "pageSource":"search",
    "hmas":[],
    "mtrs":[],
    "primarySchoolNets":[],
    "markets":[],
    "bigestAndEstate":["3-KEPPWPPAPK"],
    "phaseAndEstate":[]
}

results = []

for i in range(0, 2):
  body['offset'] = i * 24
  resp = requests.post(url, json=body)
  results.append(resp.json())
```
