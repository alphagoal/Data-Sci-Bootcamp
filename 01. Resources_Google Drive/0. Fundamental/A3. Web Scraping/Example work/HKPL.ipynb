```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text)
```

```python
books = soup.select("div.record")
```

```python
len(books)
```

```python
records = []

for book in books:

  title = book.select_one("a.title").text.strip()

  author = None

  author_tag = book.select_one("a.author")
  if author_tag is not None:
    author = author_tag.text.strip()

  records.append([title, author])
```

```python
records
```

```python
import pandas as pd
```

```python
pd.DataFrame(records)
```
