```python
import requests

url = "https://hk.centanet.com/findproperty/api/Post/Search"

result = []

for page in range(3):
  # Use this if the payload is json format
  # For centaline, use this
  response = requests.post(url, json={
      "postType":"Sale",
      "sort":"Ranking",
      "order":"Ascending",
      "size":24,
      "displayTextStyle":"WebResultList",
      "offset": page * 24,
      "pageSource":"search",
      "hmas":[],"mtrs":[],"primarySchoolNets":[],"markets":[],
      "bigestAndEstate":["3-YAPPWPPJPW"],
      "phaseAndEstate":[]
  }, headers={
      "Lang": "en"
  })

  for item in response.json()['data']:
    result.append([item['buildingName']])

  # Use this if the payload is url-encoded format, like key=value&a=b
  #requests.post(url, data={})

```
