### Batch and Epoch

For a training, we can set the **batch size** and **number of epoch**.

**Batch** is small chunk of subset of the dataset.  For example, we divided a dataset of 5000 records into batches of size 100.  In such case we will have 50 batches.

**Epoch** represents one round of training where all records in the training set is used.  In reality, we repeat the training process with multiple epoches. e.g. 10, 50, 100.
