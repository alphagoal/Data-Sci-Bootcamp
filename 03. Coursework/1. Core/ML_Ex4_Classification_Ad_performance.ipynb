
### 2.4 Logistic Regression

train test split, and train our model

**Split the data into training set and testing set using train_test_split**

````python
# Your code here
from sklearn.model_selection import train_test_split

# Daily Time Spent on Site	Age	Area Income	Daily Internet Usage	Ad Topic Line	City	Male	Country	Timestamp	Clicked on Ad

X = ad_data[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male']]
y = ad_data['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=101)

X_train
````

**Train and fit a logistic regression model on the training set.**

````python
# Your code here
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=99999)
model.fit(X_train, y_train)
pd.DataFrame(model.coef_,columns=['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male'])
````

### 2.5 Predictions and Evaluations
**Now predict values for the testing data.**

````python
# Your code here
y_pred = model.predict(X_test)
pd.DataFrame(y_test).value_counts()
````

**Check the accuracy of our preduction**

$
\frac{\text{Number of correct prediction}}{\text{Sample  count}}
$

Instruction:

1. Calculate the accuracy

````python
# Your code here
y_test == y_pred
````

````python
accuracy = (y_test == y_pred).mean()
#The definition of accuracy only applies to classification problem only. (Linear Regression uses MAE/ MSE/ RMSE/ R sq)
print(f"The accuracy of the prediction is {accuracy*100} percent.")
````

````python
from sklearn.metrics import precision_recall_fscore_support

pd.DataFrame(precision_recall_fscore_support(y_test, y_pred),index=['precision','recall','fscore','support'])
````
