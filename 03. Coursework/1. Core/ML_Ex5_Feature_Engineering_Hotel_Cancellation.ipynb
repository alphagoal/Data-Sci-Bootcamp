### 3.3 Precision, recall and F1-Score

- Preceision of X is the % correct predicted X within the group of predicted X.
- Recall of X is the % correctly predicted X within the group of originally X.
- F1 score is the harmonic mean of recall and precision.  It is the most common metric to evaluate the result of a model.
